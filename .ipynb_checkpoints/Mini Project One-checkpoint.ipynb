```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
rainfall = [12, 18, 26, 30, 45, 60]
temperature = [28, 30, 32, 35, 34, 33]
humidity = [80, 78, 75, 70, 68, 65]
weather_types = ['Sunny', 'Rainy', 'Cloudy']
weather_counts = [90, 60, 30]  # Total days in season
```

```python
total_rainall = sum(rainfall)
print(total_rainall)
```

```python
average_temperature = sum(temperature)/len(temperature)
print(average_temperature)
```

```python
max_humidity = max(humidity)
print(max_humidity)
max_index = humidity.index(max_humidity)
print(max_index)
```

```python
max_rain = max(rainfall)
print(max_rain)
max_index_rain = rainfall.index(max_rain)
print(max_index_rain)
```

```python
total_rainall = sum(rainfall)
print(total_rainall)

average_temperature = sum(temperature)/len(temperature)
print(average_temperature)

max_humidity = max(humidity)
print(max_humidity)
max_index = humidity.index(max_humidity)
print(max_index)

max_rain = max(rainfall)
print(max_rain)
max_index_rain = rainfall.index(max_rain)
print(max_index_rain)



plt.figure(figsize=(10, 8))

plt.subplot(1, 3, 1)
plt.plot(months, rainfall, marker="x", linestyle="--", color="red", linewidth=2)
plt.title("Monthly Rainfall")
plt.xlabel("Months")
plt.ylabel("Rainfall")
plt.grid(True, linestyle="--", alpha=0.6)
plt.text(months[max_index_rain],max_rain + 1, f"Peak: {max_rain} mm", ha="center", color="red")

plt.subplot(1, 3, 2)
plt.bar(months, temperature, color="pink")
plt.title("Monthly Temperature")
plt.xlabel("Months")
plt.ylabel("Temperature")

plt.subplot(1, 3, 3)
plt.pie(weather_counts,
        labels=weather_types,
        autopct="%1.1f%%",
        startangle=90)
plt.title("Weather Type Distributon")
plt.axis("equal")
plt.savefig("weather_trial.png", dpi=300)
plt.savefig("weather_trial.pdf")

plt.show()
```
