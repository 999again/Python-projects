```python
import matplotlib.pyplot as plt

# --- DATA ---
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
rainfall = [12, 18, 26, 30, 45, 60]
temperature = [28, 30, 32, 35, 34, 33]
humidity = [80, 78, 75, 70, 68, 65]
weather_types = ['Sunny', 'Rainy', 'Cloudy']
weather_counts = [90, 60, 30]

# --- CALCULATIONS ---
total_rain = sum(rainfall)
avg_temp = sum(temperature) / len(temperature)
max_rain = max(rainfall)
max_rain_index = rainfall.index(max_rain)
max_humidity = max(humidity)
max_humidity_month = months[humidity.index(max_humidity)]

print(f"Total Rainfall: {total_rain} mm")
print(f"Average Temperature: {avg_temp:.2f} Â°C")
print(f"Month with Highest Humidity: {max_humidity_month} ({max_humidity}%)")

# --- PLOTTING ---
plt.figure(figsize=(10, 8))

# Rainfall Line Chart
plt.subplot(1, 3, 1)
plt.plot(months, rainfall, marker="x", linestyle="--", color="red", linewidth=2)
plt.title("Monthly Rainfall")
plt.xlabel("Months")
plt.ylabel("Rainfall")
plt.grid(True, linestyle="--", alpha=0.6)
plt.text(months[max_rain_index], max_rain + 1, f"Peak: {max_rain} mm", ha="center", color="red")

# Temperature Bar Chart
plt.subplot(1, 3, 2)
plt.bar(months, temperature, color="pink")
plt.title("Monthly Temperature")
plt.xlabel("Months")
plt.ylabel("Temperature")

# Weather Type Pie Chart
plt.subplot(1, 3, 3)
plt.pie(weather_counts,
        labels=weather_types,
        autopct="%1.1f%%",
        startangle=90)
plt.title("Weather Type Distribution")
plt.axis("equal")

# Save figures
plt.tight_layout()
plt.savefig("weather_summary.png", dpi=300)
plt.savefig("weather_summary.pdf")
plt.show()
```
