```python
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.feature as cfeature
import cartopy.crs as ccrs

Temp_22 = xr.open_dataset("Temp22.nc")
Temp_23 = xr.open_dataset("Temp23.nc")
Temp_24 = xr.open_dataset("Temp24.nc")
```

```python
Temp_22.info()
```

```python
print("Temp_22:")
print(Temp_22)

# Preview first time slice
print("\nFirst timestep sample:")
print(Temp_22.isel(valid_time=0))

#Renaming the valid time to time
Temp_22 = Temp_22.rename({"valid_time": "time"})
Temp_23 = Temp_23.rename({"valid_time": "time"})
Temp_24 = Temp_24.rename({"valid_time": "time"})
```

```python
#combining the three years into 1 dataset
combined_temp = xr.concat([Temp_22, Temp_23, Temp_24], dim="time")
print(combined_temp)
```

```python
#slicing the location of accra from the whole ghana by using the nearest lat ang longi
Accra = combined_temp.sel(latitude = 5.56, longitude = -0.20, method="nearest")
print(Accra)
```

```python
Kumasi = combined_temp.sel(latitude = 6.69, longitude = -1.57, method="nearest")
Tamale = combined_temp.sel(latitude = 9.40, longitude = -0.84, method="nearest")
```

```python
#changing kelvin to celcius
Accra["t2m"] = Accra["t2m"] - 273.15
Kumasi["t2m"] = Kumasi["t2m"] - 273.15
Tamale["t2m"] = Tamale["t2m"] - 273.15
```

```python
#checking the shape
Accra.t2m.shape
Kumasi.t2m.shape
Tamale.t2m.shape
```

```python
#changing the xarray to a pandas dataframe
Accra_series = Accra["t2m"].to_series()
Kumasi_series = Kumasi["t2m"].to_series()
Tamale_series = Tamale["t2m"].to_series()
```

```python
#combing the three series to one data
data = pd.concat([Accra_series, Kumasi_series, Tamale_series], axis=1)
print(data)
```

```python
#Renaming the columns
data.columns = ["Accra", "Kumasi", "Tamale"]
print(data)
```

```python
print(data.head())  # Should now show 26–30 °C range for all cities
```

```python
# Make a copy of your original DataFrame
monthly_data = data.copy()

# Create a 'month' column using Period format (e.g., 2022-01, 2022-02)
monthly_data["month"] = monthly_data.index.to_period("M")
```

```python
# Group by year-month and compute monthly averages
monthly_avg = monthly_data.groupby("month")[["Accra", "Kumasi", "Tamale"]].mean()

# Preview
print(monthly_avg.head())
```

```python
monthly_avg.plot(figsize=(12, 6), marker='o')
plt.title("Monthly Average Temperature Trends (2022–2024)")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(title="City")
plt.tight_layout()
plt.show()
```

```python
# Make a copy to avoid changing your original data
yearly_data = data.copy()

# Create a 'year' column from the datetime index
yearly_data["year"] = yearly_data.index.year
```

```python
# Group by year and compute average temperature per city
yearly_avg = yearly_data.groupby("year")[["Accra", "Kumasi", "Tamale"]].mean()

# Preview the result
print(yearly_avg)
```

```python
# Plot the yearly averages
yearly_avg.plot(kind="bar", figsize=(10, 5))
plt.title("Yearly Average Temperature (2022–2024)")
plt.ylabel("Temperature (°C)")
plt.xlabel("Year")
plt.grid(axis='y')
plt.xticks(rotation=0)
plt.legend(title="City")
plt.tight_layout()
plt.show()
```
