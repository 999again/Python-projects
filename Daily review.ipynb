```python
import matplotlib.pyplot as plt
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
rainfall = [12, 18, 26, 30, 45, 60]
temperature = [28, 30, 32, 35, 34, 33]
humidity = [80, 78, 75, 70, 68, 65]
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
plt.plot(months, rainfall, marker="o", color="blue")
plt.title("Monthly Rainfall")
plt.xlabel("Months")
plt.ylabel("Rainfall")
plt.grid(True, linestyle='--', alpha=0.6)

plt.subplot(1,3, 2)
plt.bar(months, temperature, color="pink")
plt.title('Monthly Temperature')
plt.xlabel('Months')
plt.ylabel('Temperature')

plt.subplot(1, 3, 3)
plt.plot(months, humidity, marker="x", linestyle="--", color="red", linewidth=2)
plt.title("Monthly Humidity")
plt.xlabel("Months")
plt.ylabel("Humidity")
max_hum = max(humidity)
max_index = humidity.index(max_hum)
plt.text(months[max_index], max_hum + 1, f"Peak: {max_hum}%", color='red', ha='center', fontsize=9)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig("weather_subplots.png")
plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
rainfall = [12, 18, 26, 30, 45, 60]

plt.figure(figsize=(8, 5))
plt.plot(months, rainfall, marker="o", color="blue", linewidth=2)

# Customize background
plt.gca().set_facecolor("#f5f5f5")  # light grey chart area
plt.title("Monthly Rainfall", fontsize=14, fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Rainfall")
plt.grid(True, linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```
